```python
import os
import cv2
import time
```

```python
# folder names
hnm_img_final = '/content/drive/MyDrive/Temporary/CS25_Project/images_final/HnM/'
hnm_img_final_top = '/content/drive/MyDrive/Temporary/CS25_Project/images_final'

hnm_img_resized = '/content/drive/MyDrive/Temporary/CS25_Project/images_resized/HnM/'
hnm_img_resized_top = '/content/drive/MyDrive/Temporary/CS25_Project/images_resized'
```

```python
cna_img_final = '/content/drive/MyDrive/Temporary/CS25_Project/images_test/'

cna_img_resized = '/content/drive/MyDrive/Temporary/CS25_Project/images_test_resized/'
```

```python
tot_files = 0
img_files = 0
for file in os.listdir(cna_img_final):
    tot_files += 1
    if '.jpg' in file:
        img_files += 1

print(f"Total files found: {tot_files}")
print(f"Image files found: {img_files}")
```

```python
IMG_SIZE = (224, 224)

start_time = time.time()

cnt = 0

for img in os.listdir(cna_img_final):
    image = cv2.imread(f'{cna_img_final}{img}')
    try:
        imgResized = cv2.resize(image, IMG_SIZE, interpolation = cv2.INTER_AREA)
        cv2.imwrite(f'{cna_img_resized}{img}',
                    imgResized)
    # in case any images failed to be read
    except:
        print(img)
    cnt += 1
    if cnt%500 == 0:
        print(f"Completed {cnt} in {(time.time() - start_time)/60.0:.2f} mins")

execution_time = (time.time() - start_time)/60.0
print("\nresize images takes time(mins)",execution_time)
```

```python
tot_files = 0
img_files = 0
for file in os.listdir(cna_img_resized):
    tot_files += 1
    if '.jpg' in file:
        img_files += 1

print(f"Total files found: {tot_files}")
print(f"Image files found: {img_files}")
```
